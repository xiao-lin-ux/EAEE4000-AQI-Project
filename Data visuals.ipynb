For a given city, the inputs are the individual AQI values of CO, Ozone, NO2, and PM2.5, and the output is the overall AQI value. Since the data does not come with map information, for graphing, we can plot the overall AQI value for each city. Later once we have the outputs from the NN and XGBoost models we can plot those outputs as well. However, since there are over 20,000 cities, we will just pick 50 of them to graph for the visualization.

```python
from utils import *

aqi_df = df_prep(file_dir)
aqi_df
```

```python
# get a random sampling of 100 cities out of the dataset
import random
indices = random.sample(range(0,len(aqi_df)-1),50)
sample_df = aqi_df.iloc[indices]
```

```python
# plotting
sample_city_np = sample_df['City'].to_numpy() # get the city names for the x values
sample_aqi_np = sample_df['AQI Value'].to_numpy() # get the aqi values for the y values

width=10
height=5

fig = plt.figure(figsize = (width,height))
plt.bar(sample_city_np,sample_aqi_np, width=0.5)
plt.xlabel("City")
plt.xticks(rotation=90)
plt.ylabel("AQI")
plt.title("Given AQI values for 50 random cities")
plt.show()
```
